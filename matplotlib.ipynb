```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.arange(0,10)
```

```python
y = 2*x
```

```python
plt.plot(x,y);
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])

axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x,y)

axes1.plot(x,y)
```

```python
## subplots type
#plt.figure(figsize=(10,8))
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,8))

axes[0][0].plot(x,x**2)
axes[0][1].plot(y,y**2)
 
```
