```python
from matplotlib import dates
```

```python
plt.figure(figsize=(10,4),dpi=200)
ax = df['Close']['2018-01-01':'2018-03-01'].plot()

ax.xaxis.set_major_locator(dates.WeekdayLocator())
```

```python
plt.figure(figsize=(8,3),dpi=200)
ax = df['Close']['2018-01-01':'2020-03-01'].plot()

ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%Y %B'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%B'))

ax.tick_params(axis='x',which='minor',rotation=90)
ax.tick_params(axis='x',which='major',rotation=90)

plt.xticks(ha='center')

ax.yaxis.grid(True)
```

```python
#pandas rolling statistics
```

```python
df = pd.read_csv('WMT.csv',index_col='Date',parse_dates=True)
```

```python
df
```

```python
plt.figure(figsize=(10,3),dpi=200)
df['Adj Close'].plot()
```

```python
df['Adj Close'].rolling(window=3).mean() # 3 day mean
```

```python
df['Adj Close'].head()
```

```python
(65.758080 + 64.911324 + 64.704163) / 3
```

```python
plt.figure(figsize=(10,3),dpi=200)
df['Adj Close'].plot(label='adj close')
df['Adj Close'].rolling(window=200).mean().plot(label= '200 day rolling adj close')
plt.legend()
```

```python
plt.figure(figsize=(10,3),dpi=200)
df['Adj Close'].plot(label='adj close')
df['Adj Close'].rolling(window=14).std().plot(label= '14 day std rolling adj close')
plt.legend()
```

```python
plt.figure(figsize=(10,3),dpi=200)
df['Adj Close'].rolling(window=3).std().plot(label= '3 day std rolling adj close')
plt.legend()
```
